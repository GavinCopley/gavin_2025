```python
import requests,json;
url = 'https://vpic.nhtsa.dot.gov/api/vehicles/GetModelsForMakeId/440?format=json';
r = requests.get(url);
print(r.text);		
```

```python
import requests,json;
url = 'https://vpic.nhtsa.dot.gov/api/vehicles/DecodeVINValuesBatch/';
post_fields = {'format': 'json', 'data':'5XYKT3A12CG000000'};
r = requests.post(url, data=post_fields);
print(r.text);		
```

```python
import requests,json;
url = 'https://vpic.nhtsa.dot.gov/api/vehicles/DecodeVin/5UXWX7C5*BA?format=xml&modelyear=2011';
post_fields = {'format': 'json', 'data':'5XYKT3A12CG000000'};
r = requests.get(url, data=post_fields);
print(r.text);
```
